# 単一のリストまたは配列

y値となる。x値は0始まりで自動生成される。

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.plot([1, 2, 3, 4])
plt.ylabel('some numbers')
plt.show()
```

# xとyをプロット

`plot`は任意の数の引数を取る

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16],linewidth=1.0)
plt.show()
```

# プロットのスタイル

第3引数にプロットの色と線種を示す書式文字列であるオプションを指定する。書式文字列はカラー文字列と線スタイル文字列を連結する。デフォルトの書式文字列は 'b-'で、青色の実線。'ro'を指定すると赤い円でプロットする。

`axis`メソッドには`[xmin, xmax, ymin, ymax]`でxとyの範囲を指定する。

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro')
plt.axis([0, 6, 0, 20])
plt.show()
```

# 複数プロット

配列を使用して1つのコマンドで異なる書式スタイルを持つ複数の行をプロットする方法

```python
import numpy as np

# evenly sampled time at 200ms intervals
t = np.arange(0., 5., 0.2)

# red dashes, blue squares and green triangles
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

# キーワード文字列でプロットする

dataキーワード引数を指定することで文字列でプロットを生成する。

```python
print(range(5))
data = {'a': np.arange(50)}
print(data)

data['b'] = data['a'] + 10 * np.random.randn(50)
print(data)

plt.scatter('a', 'b', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

`scatter`にはx値y値とcolor、scalar(マーカーサイズ)、データを指定

```python
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

plt.scatter('a', 'b', c='c', s='d', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

```python
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]

plt.figure(1, figsize=(12, 3))

plt.subplot(131)
plt.bar(names, values)
plt.subplot(132)
plt.scatter(names, values)
plt.subplot(133)
plt.plot(names, values)
plt.suptitle('Categorical Plotting')

plt.show()
```
