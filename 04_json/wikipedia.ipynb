```python
import requests,json

keyword = 'python'
url = 'https://ja.wikipedia.org/w/api.php?format=json&action=query&prop=revisions&titles={}&rvprop=content&rvparse'.format(keyword)
response = requests.get(url)
response.raise_for_status()
data = json.loads(response.text)

print (data)
```

```python
text = ''
for pageid in data['query']['pages']:
    print (pageid)
    page = data['query']['pages'][pageid]
    text = page['revisions'][0]['*']

print (text)
```

```python
import bs4

soup = bs4.BeautifulSoup(text)
elems = soup.select('p')

print (elems)
```

```python
from IPython.display import HTML

for elem in elems:
    print(elem.getText())
```
